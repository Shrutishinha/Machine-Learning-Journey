```python
import pandas as pd
df = pd.DataFrame({
    'Student': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Marks': [70, 80, 90, 85, 75, 96.3]
})

average= df['Marks'].mean()

print("Average Marks =", average)
```

```python
P_B_by_A = 0.25
P_A = 0.7
P_B = 0.2

P_A_by_B = (P_B_by_A * P_A) / P_B

print(P_A_by_B)
```

```python
import pandas as pd
df=pd.DataFrame({10, 29, 30, 40})
even_count = (df[0] % 2 == 0).sum()
total = len(df)

p_even = even_count / total
p_odd = 1 - p_even

print("P(Even) =", p_even)
print("P(Odd) =", p_odd)
print("Sum =", p_even + p_odd)
```

```python
import pandas as pd

df1 = pd.DataFrame({'A': [1, 2, 3]})
df2 = pd.DataFrame({'B': [4, 5, 6]})

result = pd.concat([df1, df2], axis=1)

print(result)
```
