```python
# Install NLTK (if needed)
!pip install nltk scikit-learn matplotlib

import nltk

# Download required datasets (only first time)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab') # Added to resolve LookupError

# ----------------------------
# Tokenization + Stopword Removal
# ----------------------------
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

text = "NLP is fascinating!"

tokens = word_tokenize(text)

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("Filtered Tokens:", filtered_tokens)

# ----------------------------
# Lemmatization
# ----------------------------
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("better", pos='a'))
print(lemmatizer.lemmatize("running", pos='v'))
print(lemmatizer.lemmatize("mice", pos='n'))

# ----------------------------
# TF-IDF Vectorization
# ----------------------------
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

docs = ["NLP IS FUN AND EXCITING", "DATA SCIENCE IS FUN"]

matrix = tfidf.fit_transform(docs)

print("\nWords:", tfidf.get_feature_names_out())
print("\nTF-IDF Matrix:")
print(matrix.toarray())

# ----------------------------
# Visualization
# ----------------------------
import matplotlib.pyplot as plt

words = tfidf.get_feature_names_out()
scores = matrix.toarray().sum(axis=0)

sorted_data = sorted(zip(words, scores), key=lambda x: x[1])

plt.barh(
    [x[0] for x in sorted_data],
    [x[1] for x in sorted_data]
)

plt.title("TF-IDF Word Frequency")
plt.xlabel("Score")
plt.show()
```
