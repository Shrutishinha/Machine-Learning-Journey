```python
!pip install gensim
from gensim.models import Word2Vec

sentences = [
    ["nlp", "is", "great"],
    ["data", "science", "is", "fun"]
]

model = Word2Vec(
    sentences,
    vector_size=5,
    window=2,
    min_count=1
)

print(model.wv["nlp"])
```
